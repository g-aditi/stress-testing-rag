# Phoenix Crime No-RAG vs RAG Demo

Runs one incident demo using the official City of Phoenix crime dataset and prints:
- selected incident
- generated question
- answer without retrieval
- answer with retrieval

A markdown report is also saved in `outputs/`.

```python
import os

import subprocess

from pathlib import Path



project_dir = Path.cwd()

script_path = project_dir / "phoenix_crime_rag_demo.py"

python_exe = "/home/apai10/tau_env/bin/python"



if not os.getenv("OPENAI_API_KEY"):

    # Load from ~/.bashrc export line if available

    bashrc = Path.home() / ".bashrc"

    if bashrc.exists():

        for line in bashrc.read_text(encoding="utf-8", errors="ignore").splitlines():

            if line.startswith("export OPENAI_API_KEY="):

                key = line.split("=", 1)[1].strip().strip("\"'")

                if key:

                    os.environ["OPENAI_API_KEY"] = key

                break



if not os.getenv("OPENAI_API_KEY"):

    raise EnvironmentError("OPENAI_API_KEY is not set. Add it to env or ~/.bashrc")



result = subprocess.run(

    [python_exe, str(script_path)],

    cwd=str(project_dir),

    capture_output=True,

    text=True,

    check=False,

    env=os.environ.copy(),

)



print(result.stdout)

if result.returncode != 0:

    print(result.stderr)

    raise RuntimeError(f"Demo script failed with exit code {result.returncode}")



outputs_dir = project_dir / "outputs"

reports = sorted(outputs_dir.glob("phoenix_rag_demo_result_*.md"))

if reports:

    print(f"Latest report: {reports[-1]}")

else:

    print("No report found in outputs/")
```
